# Extract R_shock, V_shock

```python
def get_R_shock_V_shock(i_snapshot):
    
    ds = load_ds_from_ts(ts, snapshot_number_to_index_map[i_snapshot])
    ds.add_particle_filter("overdense")
    dd = ds.all_data()
        
    R_shock = float(np.mean(dd["overdense", "particle_position_spherical_radius"]).to("pc").value)
    V_shock = float(np.mean(dd["overdense", "particle_velocity_spherical_radius"]).to("km / s").value)
    
    return R_shock, V_shock
```

```python
def e_cool_theory(r_shock, v_shock):
    """calculates shock cooling floor in erg / Myr"""
    rho = 1.33 * m_proton
    e_dot_cool = 4*np.pi*r_shock**2 * v_shock * rho * (v_shock**2 / 2)
    return e_dot_cool * Myr
```

```python
if os.path.exists(outfile):
    df_shock = pd.read_csv(outfile)
else:
    df_shock = pd.DataFrame(
        data={
            "i_snapshot": np.empty(0, dtype=int),
            "R_shock": np.empty(0, dtype=float),
            "V_shock": np.empty(0, dtype=float),
        }
    )

for i_snapshot in df.number:
    if i_snapshot in df_shock.i_snapshot:
        print("already read snapshot {}; skipping to next".format(i_snapshot))
        continue
        
    R_shock, V_shock = get_R_shock_V_shock(i_snapshot)
    _df_shock = pd.DataFrame(
        data={
            "i_snapshot": np.array([i_snapshot], dtype=int),
            "R_shock": np.array([R_shock], dtype=float),
            "V_shock": np.array([V_shock], dtype=float),
    })

    df_shock = pd.concat([df_shock, _df_shock], ignore_index=True)
    df_shock = df_shock[["i_snapshot", "R_shock", "V_shock"]]

    df_shock.to_csv(outfile, index=False)
    print("finished reading snapshot ", i_snapshot)
    sys.stdout.flush()

df_shock["E_dot"] = e_cool_theory(df_shock.R_shock*pc, 
                                  df_shock.V_shock*km)
df_shock.to_csv(outfile, index=False)
    
```
